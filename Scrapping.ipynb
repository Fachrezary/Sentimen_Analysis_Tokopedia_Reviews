```python
from google_play_scraper import app

result = app(
    "com.tokopedia.tkpd",
    lang = "id",
    country = "id")

result
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    "com.tokopedia.tkpd",
    lang = "id", 
    country = "id",
    sort = Sort.NEWEST,
    count = 10000)

result[10000-1]
```

```python
result, _ = reviews(
    "com.tokopedia.tkpd",
    continuation_token = continuation_token) 

result
```

```python
import pandas as pd

df = pd.DataFrame(result)

csv_file_path = 'reviews_tokopedia.csv'
df.to_csv(csv_file_path, index=False, encoding='utf-8')

print(f"Data ulasan telah disimpan ke {csv_file_path}")
```
